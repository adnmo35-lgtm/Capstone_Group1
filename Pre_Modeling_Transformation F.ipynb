# Pre-Modeling Transformation Notebook


## Section 1 – Project Alignment

### Research Questions

1. What property characteristics, pricing factors, and market conditions are most associated with stale real estate listings in Edmonton?

2. How does a property’s pricing relative to its local market, including measures such as price per square foot and expectation gap, influence the likelihood of becoming a stale listing?

3. How do community-level and time-based factors, such as neighborhood trends, listing density, average days on market, year, month, and season, affect listing performance?

### Type of Analysis

The primary analytical approach for the next stage of this project is classification analysis, since the notebook below creates a binary target variable called stale_listing to distinguish between stale and non-stale listings. This allows the project to evaluate whether a property is likely to remain on the market for an extended period based on its features.

To support this classification task, the notebook includes a series of pre-modeling transformations such as:
- Data Clenaing
- Duplicate removal
- Missing value handling
- Feature engineering

Engineered variables that were created were:
- property_age
- price_per_bedroom
- price_per_sqft
- community_median_price
- expectation_gap
- community_listing_gap
- community_avg_dom

These were all created to strengthen the dataset and better capture pricing, property, and neighborhood patterns.

In addition to classification, the transformed dataset also supports descriptive analysis and segmentation, since community-level and seasonal features make it possible to compare different groups of listings and identify broader trends across the Edmonton market.
